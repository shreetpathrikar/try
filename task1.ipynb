# Bi-Variate

# as we have range of cholestrol seperating the patients by counting of  good,bad, very bad category of cholestrol 

```python

normal=[]
mildy_high=[]
high=[]
very_high=[]
for i in df["chol"]:
    if i < 150:
        normal.append(i)
    elif 150 < i < 199:
        mildy_high.append(i)
    elif 200 <= i <499:
        high.append(i)
    else:
        very_high.append(i)  
        
        
      
category_counts = {
    "Normal": len(normal),
    "Mild High": len(mildy_high),
    "High": len(high),
    "Very High": len(very_high)
}



category_labels = list(category_counts.keys())
category_values = list(category_counts.values())

plt.bar(category_labels, category_values)
plt.xlabel("cholestrol category")
plt.ylabel("number of patient")
plt.title("cholestrol patient distribution")
plt.show()


```
